mulai eda

- di listings ada simbol uang dan persentase
- di calendar juga ada, kolom available tipe boolean diubah jadi integer

```python
def clean_uang(x):
    if isinstance(x, str):
        return float(x.replace('$','').replace(',','').strip())
    return x

def clean_persen(x):
    if isinstance(x, str):
        return float(x.replace('%','').strip())
    return x
```

```python
listings['price_bersih'] = listings['price'].apply(clean_uang)
rate = ['host_response_rate', 'host_acceptance_rate']
for i in rate:
    if i in listings.columns:
        listings[i] = listings[i].apply(clean_persen)

import numpy as np

calendar['price'] = calendar['price'].astype(str).replace('nan', np.nan)
calendar['prirce_bersih'] = calendar['price'].apply(clean_uang)

calendar['is_available'] = calendar['available'].map(
    {
        't':1,
        'f':0
    }
)
```

filtering data karena:
- maksimal $13000 (150juta/malam) setelah dibersihkan
- harga minimum $1, tentu tidak ada

- mean $345 vs median $221, mean > median, outlier atas menarik rata2 secara aneh

setelah pencarian, harga termurah di SG jarang di bawah <$20, dan termahal jarang di atas $2000

```python
df_bersih = listings[
    (listings['price_bersih'] > 20) &
    (listings['price_bersih'] < 2000)
].copy()

print("Data Awal", listings.shape[0])
print("Data Awal", df_bersih.shape[0])
print("Dibuang", listings.shape[0] - df_bersih.shape[0])

print(df_bersih['price_bersih'].describe())
```

```python
sns.histplot(df_bersih['price_bersih'], bins = 50, kde = True, color = 'pink')
plt.title('Dist Harga Filter', fontsize = 14)
plt.xlabel('Harga per Malam (SGD)')
plt.show()
```
